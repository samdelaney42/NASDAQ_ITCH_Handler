```python
import struct
import sys
from pathlib import Path

sys.path.append('../src')
from event import Event
from order_book_builder import OrderBook

from itch_binary_reader import Reader
```

```python
reader = Reader()
reader.createBook("MSFT", levels=1)
reader.readData("../data/msft.bin", num_messages=1000)
book = reader.getOrderBook()
```

```python
formatted_book = book.formatBook(500)
formatted_book.head()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows = 3, cols = 1, shared_xaxes=True)

fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Bid_1, mode='lines', line=dict(color='green'), name='Best Bid'),row=1,col=1)
fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Ask_1, mode='lines', line=dict(color='red'), name='Best Ask'),row=1,col=1)

fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Bid_1_Ord, mode='lines', line=dict(color='green'), name='Bid Ords'),row=2,col=1)
fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Ask_1_Ord, mode='lines', line=dict(color='red'), name='Ask Ords'),row=2,col=1)

fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Bid_1_Vol, mode='lines', line=dict(color='green'), name='Bid Vol'),row=3,col=1)
fig.add_trace(go.Scatter(x=formatted_book.index, y=formatted_book.Ask_1_Vol, mode='lines', line=dict(color='red'), name='Ask Vol'),row=3,col=1)

fig.update_layout(title="MSFT BBO",
                  xaxis3=dict(title='Events', linecolor='black'),
                  yaxis=dict(title="Price (USD)", linecolor='black'),
                  yaxis2=dict(title='Orders @ BBO', linecolor='black'),
                  yaxis3=dict(title='Bid Ask Vol', linecolor='black'),
                  height=1000)

fig.show()
```
